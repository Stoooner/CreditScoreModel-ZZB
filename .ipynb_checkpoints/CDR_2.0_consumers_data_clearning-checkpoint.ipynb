# 1. 直租2.0用户信息预处理
### 1.1 获取CDR_2.0_Consumers_Info.xlsx中所有直租2.0用户的信息

```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
data = pd.read_excel('Original_data/CDR_2.0_Consumers_Info.xlsx')
# report = pp.ProfileReport(data)
# report.to_file('CDR_2.0_Consumers_Info_report.html')
```

```python
data.describe()
```

通过对数据的统计和观察可以发现：
>1. 很多的特征都是整列全为空值的情况，所以将其手动删除掉；
>2. 总的数据量为652个，其中数据部分特征存在超级多的信息缺失，为不影响后续分析，将缺失较多的特征进行删除，其中缺失较多的特征及其对应的数据量为：
| 缺失特征      |    缺失数目 | 缺失比例  |
| :-------- | --------:| :--: |
|办公电话 | 569 |  87.3%   |
|房产性质 | 542 |  83.1%  |
|自有房产 |  369 | 56.6%  |
|车牌 |  614 | 94.2%  |

```python
data.drop(columns = ['办公电话','房产性质','自有房产','车牌'], inplace=True)
```

### 1.2 各个特征分析
#### 1.2.1 民族特征分析

```python
data['民族'].value_counts()
# 从下面的汇总可以看出，数据中总共的民族包含有14个，接下来将各个民族的数据合并，然后进行数值化
```

```python
# 定义一个包含数据中所有名族的字典，方便进行合并与隐射
ethnicity_enumerate = {'汉':0, '汉族':0, '汗':0, '苗':1, '苗族':1, '满':2, '满族':2,
                      '土家':3, '土家族':3, '布依':4, '布依族':4, '侗':5, '侗族':5,
                      '壮':6, '壮族':6, '彝':7, '彝族':7, '瑶':8, '瑶族':8, 
                      '土':9, '回':10, '回族':10, '羌':11, '羌族':11, '畲':12, '畲族':12,
                      '白族':13, '黎':14,'黎族':14}
data = pd.read_excel('Original_data/CDR_2.0_Consumers_Info.xlsx')
# 将每个人的民族的字段合并和映射到数字，方便后面进行分析
data['民族'] = pd.Series(list(map(lambda x: ethnicity_enumerate[x], data['民族'])))
data.to_excel('Original_data/CDR_2.0_Consumers_Info.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt
# a = np.array(data['民族'].value_counts())
ethnicities_nums = list(data['民族'].value_counts().index)
ethnicities_size = list(data['民族'].value_counts())
# 用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei'] 

labels = ['汉族','满族','布依','苗族','土家','彝族','壮族',
         '瑶族','侗族','羌族','回族','土族','黎族','白族','畲族']
explode = (0,0,0,0.1,0,0)
plt.pie(ethnicities_size, labels=labels, autopct='%1.1f%%', shadow=False, startangle=150)
plt.title("各个民族的人数")
plt.show()  
```

```python
for i in range(len(labels)):
    plt.bar(labels[i], ethnicities_size[i])
    plt.text(labels[i], ethnicities_size[i], '%i' % ethnicities_size[i], ha='center', va='bottom')
plt.title('各个民族的人数')
plt.xlabel('民族')
plt.ylabel('人数')
plt.savefig('各个民族的人数.jpg')
```

### 1.3 变量相关性分析

```python
# 相关性分析：
corr = data.corr()#计算各变量的相关性系数
# # x轴标签
xticks = ['x0','x1','x2','x3','x4','x5','x6','x7','x8','x9','x10']
xticks = ['性别', '年龄', '民族', '电话号码', '本地居住时长', '婚姻状况', '教育程度', '现住房性质',
          '有无驾照', '自有车辆', '现单位工作时长', '职务', '月收入', '年收入']
# y轴标签
yticks = list(corr.index)
fig = plt.figure(figsize=(14,12))
ax1 = fig.add_subplot(1, 1, 1)
sb.heatmap(corr, annot=True, cmap='rainbow', axv=vax1, linewidths=.5,
           annot_kws={'size': 12, 'weight': 'bold', 'color': 'black'})#绘制相关性系数热力图
ax1.set_xticklabels(xticks, rotation=-45, fontsize=14)
ax1.set_yticklabels(yticks, rotation=0, fontsize=14)
plt.savefig('矩阵热力图.png', dpi=200)
plt.show()
corr
```

```python
import random
from pyecharts import heatmap

x_axis = [
    "12a", "1a", "2a", "3a", "4a", "5a", "6a", "7a", "8a", "9a", "10a", "11a",
    "12p", "1p", "2p", "3p", "4p", "5p", "6p", "7p", "8p", "9p", "10p", "11p"]
y_axis = [
    "Saturday", "Friday", "Thursday", "Wednesday", "Tuesday", "Monday", "Sunday"]
data = [[i, j, random.randint(0, 50)] for i in range(24) for j in range(7)]
heatmap_ = heatmap()
heatmap_.add(
    "热力图直角坐标系",
    x_axis,
    y_axis,
    data,
    is_visualmap=True,
    visual_text_color="#000",
    visual_orient="horizontal",
)
heatmap_.render()
```
