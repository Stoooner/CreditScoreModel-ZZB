# 订单数据_被拒件数据处理

```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
data = pd.read_excel('Original_data_2/订单数据-被拒件.xlsx', header=None, index=False)
# report = pp.ProfileReport(data)
# report.to_file('订单数据_被拒件数据报告.html')
```

```python
# data.drop(index=[0], inplace=True)
data.head()
```

```python
permanent_address = list(data.iloc[:,6][1:])
permanent_address
```

```python
import cpca
permanent_address = cpca.transform(permanent_address)
permanent_address
```

```python
# 用来正常显示中文标签
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] 
provinces = []
numbers = []
for key, value in dict(permanent_address['省'].value_counts()).items():
    provinces.append(key)
    numbers.append(value)
plt.figure(figsize=(20,10))
for i in range(len(provinces)):
    plt.bar(provinces[i], numbers[i])
    plt.text(provinces[i], numbers[i]+0.05, '%i' % numbers[i], ha='center', va='bottom', fontsize=15)
plt.xticks(np.arange(len(provinces)), provinces, rotation=-55)#rotation控制倾斜角度
plt.xticks(fontsize=13)
plt.yticks(fontsize=15)
plt.title('订单数据_被拒件用户的户籍省情况', fontsize=16)
plt.ylabel('人数', fontsize=15)
plt.savefig('订单数据_被拒件用户的户籍省情况.jpg')
```
