# 用户风险报告数据预处理
## 1. 数据读取

```python
import pandas as pd
import numpy as np
data = pd.read_excel('Original_data/risk_customer_report.xlsx', index_col=0, header=0)
data
```

## 2. 运营商信息读取分析

```python
# 首先将所有运营商信息从字符串转换为列表(eval)
operators = []
for i in range(data['运营商信息'].shape[0]):
    operators.append(eval(data['运营商信息'][i]))
```

```python
# 将每个人的运营商信息从列表中的字典的形式转换为DataFrame
import sys
# 创建一个新的DataFrame, 包含的columns有['attribution', 'costLevel', 'dataSource', 'isSame', 'netTime', 'bankAccountCheck']
operators_dataframe = pd.DataFrame(columns=['attribution', 'costLevel', 'dataSource', 'isSame', 
                                   'netTime', 'bankAccountCheck'])
for i in range(len(operators)):
    # 循环的将每个人的运行商信息(2个)合并到operators_dataframe中
    temp = pd.DataFrame(operators[i])
    operators_dataframe = pd.concat([operators_dataframe, temp], sort=False)
```

```python
operators_dataframe
```

```python
eval(data['个人股权a'][0])['blackLists'][0]
```

```python
# 将所有个人股权信息从字符串转换为列表(eval)
personal_equity = []
for i in range(data['个人股权'].shape[0]):
    personal_equity.append(eval(data['个人股权'][i]))
```

```python
personal_equity[0]
```

```python
personal_equity_dataframe = pd.DataFrame(columns=['dataSource', 'errorMessage', 'investmentCount', 'investmentTotalMoney', 
                                                  'items', 'subReportType', 'subReportTypeCost', 'treatResult'])
for i in range(len(personal_equity)):
    temp = pd.DataFrame([personal_equity[i]])
    personal_equity_dataframe = pd.concat([personal_equity_dataframe, temp], sort=False)
personal_equity_dataframe.loc[0, 'errorMessage'] = np.nan
```

```python
# pd.DataFrame(personal_equity_dataframe['items'].iloc[0])
personal_equity_item = []
for i in range(personal_equity_dataframe['items'].shape[0]):
    personal_equity_item.append(personal_equity_dataframe['items'].iloc[i])
```

```python
personal_equity_item_dataframe = pd.DataFrame(columns=['contributiveFund', 'contributivePercent', 'corpName', 'corpTypeCaption', 
                                                       'corpTypeID', 'name', 'registFund', 'registNo', 'statusCaption', 'statusID'])

for i in range(len(personal_equity_item)):
    if type(personal_equity_item[i]) == list:
        temp = pd.DataFrame([personal_equity_item[i][0]])
    else:
        temp = pd.DataFrame([personal_equity_item[i]])
    personal_equity_item_dataframe = pd.concat([personal_equity_item_dataframe, temp], sort=False)
# personal_equity_item[0]
```

```python
personal_equity_item_dataframe.drop(0,axis=1,inplace=True)
```

```python
# 创建一个新的DataFrame用于将包含字典的列表转换为Dataframe
black_list_dataframe = pd.DataFrame(columns=['dataSource', 'hasCheated', 'hasOverdue', 'isEconnoisseur', 
                                            'hasExecution', 'isUnpromising'])
for i in range(len(black_list)):
    # 循环的将每个人的运行商信息(2个)合并到operators_dataframe中
    temp = pd.DataFrame(black_list[i])
    black_list_dataframe = pd.concat([black_list_dataframe, temp], sort=False)
```
