# 订单数据_已完成数据集分析

```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
data = pd.read_csv('Original_data_2/订单数据_已完成.csv')
```

```python
# 用来正常显示中文标签
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] 

permanent_address = pd.read_csv('Original_data_2/户籍信息.csv')
provinces = []
numbers = []
for key, value in dict(permanent_address['省'].value_counts()).items():
    provinces.append(key)
    numbers.append(value)
plt.figure(figsize=(20,10))
for i in range(len(provinces)):
    plt.bar(provinces[i], numbers[i])
    plt.text(provinces[i], numbers[i]+0.8, '%i' % numbers[i], ha='center', va='bottom', fontsize=15)
plt.xticks(np.arange(len(provinces)), provinces, rotation=-55)#rotation控制倾斜角度
plt.xticks(fontsize=13)
plt.yticks(fontsize=15)
plt.title('订单数据_已完成用户的户籍省情况', fontsize=16)
plt.ylabel('人数', fontsize=15)
plt.savefig('订单数据_已完成用户的户籍省情况.jpg')
```

# 2. 颜色数据情况

```python
color_types = []
color_numbers = []
for key, value in dict(data['颜色'].value_counts()).items():
    color_types.append(key)
    color_numbers.append(value)
plt.figure(figsize=(20,10))
for i in range(len(color_types)):
    plt.bar(color_types[i], color_numbers[i])
    plt.text(color_types[i], color_numbers[i]+0.5, '%i' % color_numbers[i], ha='center', va='bottom', fontsize=15)
plt.xticks(np.arange(len(color_types)), color_types, rotation=-90)#rotation控制倾斜角度
plt.xticks(fontsize=13)
plt.yticks(fontsize=15)
plt.title('订单数据_已完成用户的车辆颜色', fontsize=16)
plt.ylabel('颜色数目', fontsize=15)
plt.savefig('订单数据_已完成用户的颜色数目情况.jpg')
```

```python
ID_card = np.array(data['承租人身份证号'])
ID_card_info_7_14 = []
for i in range(ID_card.shape[0]):
    ID_card_info_7_14.append(ID_card[i][6:10] + 
                             '-' + ID_card[i][10:12] + 
                             '-' + ID_card[i][12:14])
# ID_card_info_7_14
data['出生年月'] = pd.Series(ID_card_info_7_14)
```

```python
import datetime as dt
years = []
age = []
for i in range(len(ID_card_info_7_14)):
    years.append(int(ID_card_info_7_14[i].split('-')[0]))
years = pd.Series(years)
age = dt.datetime.today().year - years
# age
data['年龄'] = pd.Series(age)
```

```python
# 画出年龄的频率直方图
import seaborn as sb

sb.set(style="darkgrid")
plt.hist(age, density=True)
# plt.figure(figsize=(15, 10))
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('The Histgram of Age')
plt.savefig('订单数据_已完成用户年龄.jpg')
```

# 3. 车辆品牌

```python
car_types = []
car_numbers = []
plt.rcParams['font.sans-serif']=['SimHei'] #显示中文标签
for key, value in dict(data['车辆品牌'].value_counts()).items():
    car_types.append(key)
    car_numbers.append(value)
plt.figure(figsize=(20,10))
for i in range(len(car_types)):
    plt.bar(car_types[i], car_numbers[i])
    plt.text(car_types[i], car_numbers[i]+0.5, '%i' % car_numbers[i], ha='center', va='bottom', fontsize=15)
plt.xticks(np.arange(len(car_types)), car_types, rotation=-90)#rotation控制倾斜角度
plt.xticks(fontsize=13)
plt.yticks(fontsize=15)
plt.title('订单数据_已完成用户的车辆品牌', fontsize=16)
plt.ylabel('品牌数目', fontsize=15)
plt.savefig('订单数据_已完成用户的车辆品牌数目情况.jpg')
```

# 4. 车辆指导价

```python
import seaborn as sb
prices = [float(i.split('万')[0]) for i in list(data['车辆指导价'])]
sb.set(style="darkgrid")
plt.hist(prices, bins=50, density=True)
plt.xlabel('prices')
plt.ylabel('frequency')
plt.title('The Histgram of Car Prices')
plt.savefig('订单数据_已完成用户车辆指导价格.jpg')
```

```python
# 连续数据分段
bins = [5, 10, 15, 20, 25, 30, 35, 40, 45]
cut1 = pd.cut(prices, bins)
cut1.value_counts()
keys = ['(5,10]','(10,15]','(15,20]','(20,25]',
        '(25,30]','(30,35]','(35,40]','(40,45]']
values = [170, 151, 16, 7, 4, 1, 0, 1]
plt.rcParams['font.sans-serif']=['SimHei'] #显示中文标签
plt.figure(figsize=(20,10))
for i in range(len(values)):
    plt.bar(keys[i], values[i])
    plt.text(keys[i], values[i]+0.8, '%i' % values[i], ha='center', va='bottom', fontsize=15)
plt.xticks(np.arange(len(keys)), keys, rotation=-45)#rotation控制倾斜角度
plt.xticks(fontsize=18)
plt.yticks(fontsize=15)
plt.title('订单数据_已完成用户的车辆指导价', fontsize=16)
plt.ylabel('人数', fontsize=15)
plt.savefig('订单数据_已完成用户的车辆指导价情况.jpg')
```
