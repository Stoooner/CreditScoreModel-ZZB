## 1. 直租1.0用户信息预处理
### 1.1 获取CDR_1.0_Consumers_Info.xlsx中所有直租1.0用户的信息

```python
import pandas as pd
import numpy as np
data = pd.read_excel('Original_data/CDR_1.0_Consumers_Info.xlsx')
data
```

## 1.2 用户性别补全
从数据中可以看出，从饶光伦这条数据开始，用户的性别和出生年月信息是空白的，但是由于用户的身份证信息已经包含了性别和出生年月的信息，因此可以直接通过身份证信息将缺失的信息不全。其中身份证性别区分要看第17位数字,第17位数字代表性别,奇数为男,偶数为女。
18位身份证号码各位的含义:
>1. 1-2位省、自治区、直辖市代码；
>2. 3-4位地级市、盟、自治州代码；
>3. 5-6位县、县级市、区代码；
>4. 7-14位出生年月日，比如19670401代表1967年4月1日；
>5. 15-17位为顺序号，其中17位（倒数第二位）男为单数，女为双数；

```python
# get 17th num of id cards of gender lose from CDR 1.0 consumers info
# get start index of gender lose from CDR 1.0 consumers info
gender_lose_start_index = data[data['性别'].isnull()].index[0]
# get id cards of gender lose of CDR 1.0 consumers
CDR_1_id_card_info = np.array(data.loc[gender_lose_start_index:,:]['身份证'])
print("ID Card information of CDR 1.0 of Consumers:\n",CDR_1_id_card_info.shape)
# store the 17th number of id card of gender lose CDR 1.0
CDR_1_id_card_info_17 = []
for i in range(CDR_1_id_card_info.shape[0]):
    if int(CDR_1_id_card_info[i][16]) % 2 == 0:
        CDR_1_id_card_info_17.append(['女',int(CDR_1_id_card_info[i][16])])
    else:
        CDR_1_id_card_info_17.append(['男',int(CDR_1_id_card_info[i][16])])
CDR_1_id_card_info_17 = np.array(CDR_1_id_card_info_17)
```

```python
# 获取到性别和出生年月缺失的直租1.0用户的身份证第17位之后, 以及对应的性别之后，将信息写入到CDR_1.0_Consumers_Info.xlsx
temp_1 = [i for i in range(gender_lose_start_index)]
# 这里是一个需要注意的点，如果需要用一个Series来填充nan值，则temp的长度要和整个性别列的长度一致，否则不能进行填充
temp_1.extend(list(CDR_1_id_card_info_17[:,0]))
temp_1 = pd.Series(temp_1)
data['性别'].fillna(temp_1, inplace=True)
# 统计性别这一列中男女两类元素各自的个数
data['性别'].value_counts()
```

## 1.3 用户出生日期补全

```python
birthday_lose_start_index = data[data['出生年月日'].isnull()].index[0]
CDR_1_id_card_info_birthday = np.array(data.loc[birthday_lose_start_index:,:]['身份证'])
print("ID Card information of CDR 1.0 of Consumers:\n",CDR_1_id_card_info_birthday.shape)
CDR_1_id_card_info_7_14 = []
for i in range(CDR_1_id_card_info_birthday.shape[0]):
    CDR_1_id_card_info_7_14.append(CDR_1_id_card_info_birthday[i][6:10] + \
                                   '-' + CDR_1_id_card_info_birthday[i][10:12] + \
                                   '-' + CDR_1_id_card_info_birthday[i][12:14])
```

```python
# 获取到性别和出生年月缺失的直租1.0用户的身份证第17位之后, 以及对应的性别之后，将信息写入到CDR_1.0_Consumers_Info.xlsx
temp_2 = [i for i in range(birthday_lose_start_index)]
# 这里是一个需要注意的点，如果需要用一个Series来填充nan值，则temp的长度要和整个性别列的长度一致，否则不能进行填充
temp_2.extend(CDR_1_id_card_info_7_14)
temp_2 = pd.Series(temp_2)
data['出生年月日'].fillna(temp_2, inplace=True)
# 将性别和出生年月日都缺失的人员填充完毕之后，将data重新写入到源文件
data.to_excel('Original_data/CDR_1.0_Consumers_Info.xlsx', index=False)
```

## 1.5 重复值删除

```python
# 统计姓名这一列
data['姓名'].value_counts()
```

```python
# df.duplicated()用于判断数据中是否有重复数据，duplicated中给定的是一个包含列名关键字的数组，这样的话就会判断两个数据见所有的关键字
# data[data.duplicated(['姓名', '身份证', '电话号码', '银行卡号', '性别', '出生年月日'])].reset_index(drop=True)
#  可以看到索引乱了,我们使用data.reset_index(),里面的参数drop=True，表明要舍掉原来的索引，不然的话原来的索引会保留下来。
data[data.duplicated()].reset_index(drop=True)
```

```python
# 去除指定列名下有重复的数据
data.drop_duplicates(inplace=True)
# 去除完毕所有的重复数据之后，将更新的数据复写到CDR_1.0_Consumers_Info_new.xlsx中
data.to_excel('Original_data/CDR_1.0_Consumers_Info.xlsx', index=False)
```

>在去除掉所有特征都相同的数据之后，发现还有部分数据存在个别特征相同的情况，例如：姓名相同，但别的特征全都不同的数据，或者某些特征相同，但是另一些特征全都不同的情况。下面对这样情况的数据进行清除，采取的清除规则是以唯一识别特征：身份证为基础进行清除。此外，如果身份证相同，且银行卡号也相同的数据，则进行清除。

```python
# 统计身份证和银行卡号相同的两列这一列
data[data.duplicated(['身份证', '银行卡号'])].reset_index(drop=True)
```

```python
data.drop_duplicates(['身份证', '银行卡号'],inplace=True)
data.to_excel('Original_data/CDR_1.0_Consumers_Info.xlsx', index=False)
```

```python
# 最后再判断是否有缺失值
data.isnull().any()
```

## 1.6 判断出生年月是否有异常值

```python
# 从用户的出生年月日信息推出用户的年龄
import datetime as dt
# 当前的年份
# v = list(map(lambda x,y:x - y,a,b))
now_year = dt.datetime.today().year 
consumers_bir_years = pd.to_datetime(data['出生年月日'], format='%Y-%m-%d')
data['年龄'] = now_year - consumers_bir_years.dt.year
data.to_excel('Original_data/CDR_1.0_Consumers_Info.xlsx', index=False)
```

```python
# 根据用户的年龄制作箱线图观察年龄数据中是否存在异常值：
import matplotlib.pyplot as plt
import seaborn as sb
plt.boxplot(data.iloc[:,6])
plt.title('Boxplot of Age')
data.iloc[:,6].describe()
```

```python
# 画出年龄的频率直方图
sb.set(style="darkgrid")
plt.hist(data['年龄'],bins=43, density=True)
# plt.figure(figsize=(15, 10))
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('The Histgram of Age')
plt.savefig('age hist.jpg')
print(data[data['年龄']>=60]['年龄'].shape)
```

根据实际的讨论，得出的意见是将60作为阈值，将年龄超过60的数据给筛掉。

```python
# 将数据中年龄大于60的数据给清除掉
data = data[data['年龄'] <= 60]
data.to_excel('Original_data/CDR_1.0_Consumers_Info.xlsx', index=False)
data['年龄'].describe()
sb.distplot(data['年龄'], color="g", bins=40)
```

观察男女比例：
